```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "animals.csv"
animals_data_df = pd.read_csv(csv_file)
animals_data_df.head()
```

```python
animals_data_df.rename(columns = {'Animal Name':'animal_name'}, inplace = True)
animals_data_df.head()
```

```python
new_animals_data_df = animals_data_df[['animal_name', 'Kingdom', 'Phylum', 'Class']].copy()
new_animals_data_df.head()
```

```python
 csv_file = "zoo2.csv"
zoo_data_df = pd.read_csv(csv_file)
zoo_data_df.head()
```

```python
new_zoo_data_df = zoo_data_df[['animal_name', 'hair', 'feathers', "airborne", "breathes", "legs", "aquatic", "backbone"]].copy()
new_zoo_data_df.head()
```

```python
#delete NA's
new_animals_data_df.dropna(inplace = True)
new_zoo_data_df.dropna(inplace = True)
```

```python
protocol = 'postgresql'
username = '<username>'
password = '<password>'
host = 'localhost'
port = 5432
database_name = 'zooanimals_db'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
#load into Postgres
new_animals_data_df.to_sql(name='animals', con=engine, if_exists='append', index=False)
```

```python
#load intro Postgres
new_zoo_data_df.to_sql(name='zoo', con=engine, if_exists='append', index=False)
```
