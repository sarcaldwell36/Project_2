```python
#Loading in dependancies 

import pandas as pd
from sqlalchemy import create_engine
```

```python
#Reading in the CSV file

csv_file = "animals.csv"
animals_data_df = pd.read_csv(csv_file)
animals_data_df.head()
```

```python
#Transforming

animals_data_df.rename(columns = {'Animal Name':'animal_name'}, inplace = True)
animals_data_df.head()
```

```python
#Narrowing down the data set

new_animals_data_df = animals_data_df[['animal_name', 'Kingdom', 'Phylum', 'Class']].copy()
new_animals_data_df.head()
```

```python
#Reading in the json

json_file = "zoo.json"
zoo_data_df = pd.read_json(json_file, typ='series')
zoo_data_df.head()
```

```python
#Transforming the json to a data frame

type(zoo_data_df)
zoo_data_df = pd.DataFrame(zoo_data_df)
```

```python
#Resetting the index

zoo_data_df.reset_index(inplace=True)
```

```python
#Creating column names

column_names=["scientific_name","animal_name"]
zoo_data_df.columns = column_names
zoo_data_df.head()
```

```python
#Delete NA's
new_animals_data_df.dropna(inplace = True)
zoo_data_df.dropna(inplace = True)
```
